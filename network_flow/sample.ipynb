## 最大流問題を解く

```python
from maxflow import maxflow, visualize
```

ネットワークを構築する

```python
f = maxflow(9)
f.add_random_edge(15)
```

ビジュアライズの準備

```python
vis = visualize(f)
```

ネットワークの確認

```python
vis.show_graph("capacity")
```

アルゴリズム終了まで以下のセルを繰り返し実行する

残余グラフを更新して表示

```python
vis.update_residual()
vis.show_graph("residual")
```

増加路を表示する

```python
vis.show_path()
```

flowを更新して表示する

```python
vis.update_flow()
```

上のセルがNo updateになったら以下のセルを実行する

```python
print(f"max flow is {vis.maxflow.sum_f()}")
```
